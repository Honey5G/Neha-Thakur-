Question 1: What is hypothesis testing in statistics?

Ans: Hypothesis testing is a statistical method used to make inferences or draw conclusions about a population parameter based on a sample of data. It involves formulating a null hypothesis (H0) and an alternative hypothesis (Ha), collecting data, and using statistical tests to determine whether there is enough evidence to reject the null hypothesis in favor of the alternative hypothesis. The process typically involves calculating a test statistic and comparing it to a critical value or p-value to make a decision about the hypotheses.

Question.2 What is the null hypothesis, and how does it differ from the alternative
hypothesis?


Ans: The null hypothesis ($H_0$$H_0$) is a statement of no effect or no difference. It is the hypothesis that the researcher is trying to disprove. The alternative hypothesis ($H_a$$H_a$ or $H_1$$H_1$) is a statement that contradicts the null hypothesis. It is the hypothesis that the researcher is trying to prove.

For example, if a researcher is testing whether a new drug lowers blood pressure, the null hypothesis would be that the drug has no effect on blood pressure ($H_0$$H_0$: The drug does not lower blood pressure). The alternative hypothesis would be that the drug does lower blood pressure ($H_a$$H_a$: The drug lowers blood pressure)

Ques.3 Explain the significance level in hypothesis testing and its role in deciding the outcome of a test.

Ans: The significance level ($\alpha$$\alpha$) in hypothesis testing is the probability of rejecting the null hypothesis when it is actually true. It is the threshold for determining statistical significance. Common significance levels are 0.05 (5%), 0.01 (1%), and 0.10 (10%).

The role of the significance level in deciding the outcome of a test is as follows:

Comparing p-value to $\alpha$$\alpha$: After performing a statistical test, you obtain a p-value. The p-value is the probability of observing the data (or more extreme data) if the null hypothesis were true. You compare this p-value to the predetermined significance level ($\alpha$$\alpha$).
Decision Rule:


If the p-value is less than or equal to $\alpha$$\alpha$ (p-value $\le \alpha$$\le \alpha$), you reject the null hypothesis. This means that the observed data is unlikely to have occurred if the null hypothesis were true, providing enough evidence to support the alternative hypothesis.
If the p-value is greater than $\alpha$$\alpha$ (p-value $> \alpha$$> \alpha$), you fail to reject the null hypothesis. This means that the observed data is reasonably likely to have occurred if the null hypothesis were true, and there is not enough evidence to support the alternative hypothesis.

Ques.4 What are Type I and Type II errors? Give examples of each.


Ans: In hypothesis testing, there are two main types of errors you can make:

Type I Error (False Positive): This occurs when you reject the null hypothesis ($H_0$$H_0$) when it is actually true. In other words, you conclude there is a significant effect or difference when there isn't one. The probability of making a Type I error is equal to the significance level ($\alpha$$\alpha$).

Example: A medical test for a disease gives a positive result (rejecting the null hypothesis that the person is healthy) when the person does not actually have the disease (the null hypothesis is true).
Type II Error (False Negative): This occurs when you fail to reject the null hypothesis ($H_0$$H_0$) when it is actually false. In other words, you conclude there is no significant effect or difference when there actually is one. The probability of making a Type II error is denoted by $\beta$$\beta$.

Example: A medical test for a disease gives a negative result (failing to reject the null hypothesis that the person is healthy) when the person actually has the disease (the null hypothesis is false).

Ques.5 : What is the difference between a Z-test and a T-test? Explain when to use
each.

Ans:
The Z-test and the T-test are both statistical hypothesis tests used to determine if there is a significant difference between the means of two groups or between a sample mean and a population mean. The key difference lies in whether the population standard deviation is known or unknown, and the sample size.

Here's a breakdown of the differences and when to use each:

Z-test:

When to use:
When the population standard deviation ($\sigma$$\sigma$) is known.
When the sample size is large (generally considered $n \ge 30$$n \ge 30$), even if the population standard deviation is unknown (due to the Central Limit Theorem, the sample standard deviation can be a good estimate of the population standard deviation in large samples).
When the data is normally distributed.
Characteristics: It uses the Z-distribution (standard normal distribution) to calculate the critical values or p-values.
T-test:

When to use:
When the population standard deviation ($\sigma$$\sigma$) is unknown.
When the sample size is small (generally considered $n < 30$$n < 30$).
When the data is approximately normally distributed (though the t-test is relatively robust to minor deviations from normality, especially with larger sample sizes).
Characteristics: It uses the T-distribution, which has heavier tails than the Z-distribution. The shape of the T-distribution depends on the degrees of freedom, which are related to the sample size. As the sample size increases, the T-distribution approaches the Z-distribution.
In summary:

Feature	Z-test	T-test
Population $\sigma$$\sigma$	Known	Unknown
Sample Size	Large ($n \ge 30$$n \ge 30$)	Small ($n < 30$$n < 30$)
Distribution Used	Z-distribution (Normal)	T-distribution

Ques.6 Write a Python program to generate a binomial distribution with n=10 and
p=0.5, then plot its histogram.


```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# Parameters for the binomial distribution
n = 10  # number of trials
p = 0.5 # probability of success on each trial
size = 1000 # number of experiments

# Generate the binomial distribution
binomial_data = np.random.binomial(n, p, size)

# Plot the histogram
plt.figure(figsize=(10, 6))
sns.histplot(binomial_data, bins=np.arange(-0.5, n + 1.5, 1), kde=False, color='skyblue', stat='density')
plt.title('Histogram of Binomial Distribution (n=10, p=0.5)')
plt.xlabel('Number of successes')
plt.ylabel('Probability Density')
plt.xticks(range(n + 1))
plt.grid(axis='y', alpha=0.75)
plt.show()
```

Ques.8 : Write a Python script to simulate data from a normal distribution and
calculate the 95% confidence interval for its mean. Plot the data using Matplotlib.

```python
import numpy as np
import matplotlib.pyplot as plt
import scipy.stats as stats

# Parameters for the normal distribution
mu = 50  # mean
sigma = 5 # standard deviation
size = 200 # number of data points

# Simulate data from a normal distribution
np.random.seed(42) # for reproducibility
simulated_data = np.random.normal(mu, sigma, size)

# Calculate the 95% confidence interval for the mean
confidence_level = 0.95
degrees_freedom = size - 1

# Calculate the sample mean and standard error
sample_mean = np.mean(simulated_data)
standard_error = stats.sem(simulated_data) # Standard error of the mean

# Calculate the confidence interval using the t-distribution
confidence_interval = stats.t.interval(confidence_level, degrees_freedom,
                                       loc=sample_mean, scale=standard_error)

print(f"Simulated Data Mean: {sample_mean:.4f}")
print(f"95% Confidence Interval for the Mean: ({confidence_interval[0]:.4f}, {confidence_interval[1]:.4f})")

# Plot the data
plt.figure(figsize=(10, 6))
plt.hist(simulated_data, bins=20, density=True, alpha=0.6, color='g')

# Plot the theoretical normal distribution
xmin, xmax = plt.xlim()
x = np.linspace(xmin, xmax, 100)
p = stats.norm.pdf(x, mu, sigma)
plt.plot(x, p, 'k', linewidth=2)

plt.title('Histogram of Simulated Normal Distribution and 95% Confidence Interval')
plt.xlabel('Value')
plt.ylabel('Density')

# Add a vertical line for the sample mean
plt.axvline(sample_mean, color='red', linestyle='dashed', linewidth=2, label=f'Sample Mean: {sample_mean:.2f}')

# Add lines for the confidence interval
plt.axvline(confidence_interval[0], color='purple', linestyle='dashed', linewidth=2, label=f'95% CI: ({confidence_interval[0]:.2f}, {confidence_interval[1]:.2f})')
plt.axvline(confidence_interval[1], color='purple', linestyle='dashed', linewidth=2)

plt.legend()
plt.grid(True, alpha=0.6)
plt.show()
```

Ques.9 : Write a Python function to calculate the Z-scores from a dataset and
visualize the standardized data using a histogram. Explain what the Z-scores represent
in terms of standard deviations from the mean.
(Include your Python code and output in the code box below.)

```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

def calculate_z_scores(data):
  """
  Calculates the Z-scores for a given dataset.

  Args:
    data: A list or numpy array of numerical data.

  Returns:
    A numpy array of Z-scores.
  """
  data_mean = np.mean(data)
  data_std = np.std(data)
  z_scores = (data - data_mean) / data_std
  return z_scores

# Example usage with the sample data from Ques.7
sample_data = [49.1, 50.2, 51.0, 48.7, 50.5, 49.8, 50.3, 50.7, 50.2, 49.6,
               50.1, 49.9, 50.8, 50.4, 48.9, 50.6, 50.0, 49.7, 50.2, 49.5,
               50.1, 50.3, 50.4, 50.5, 50.0, 50.7, 49.3, 49.8, 50.2, 50.9,
               50.3, 50.4, 50.0, 49.7, 50.5, 49.9]

z_scores_data = calculate_z_scores(sample_data)

print("Sample Data Z-scores:")
print(z_scores_data)

# Visualize the standardized data using a histogram
plt.figure(figsize=(10, 6))
sns.histplot(z_scores_data, bins=20, kde=True, color='lightcoral')
plt.title('Histogram of Z-scores')
plt.xlabel('Z-score')
plt.ylabel('Frequency')
plt.grid(axis='y', alpha=0.75)
plt.show()
```

**Explanation of Z-scores:**

A Z-score (also known as a standard score) tells you how many standard deviations away from the mean a particular data point is.

*   A Z-score of 0 means the data point is exactly at the mean.
*   A positive Z-score means the data point is above the mean. For example, a Z-score of 1 means the data point is one standard deviation above the mean.
*   A negative Z-score means the data point is below the mean. For example, a Z-score of -1 means the data point is one standard deviation below the mean.

Z-scores allow you to standardize data from different distributions, making it easier to compare values that would otherwise be on different scales. The histogram of the Z-scores will show you the distribution of the data after it has been standardized to have a mean of 0 and a standard deviation of 1.
